#### Correlación entre variables

Existen relaciones lineales entre algunas variables relacionadas con la entrega de créditos. ¿Qué tan fuerte es la correlación de los ingresos del solicitante con el monto del préstamo?

1. Gráficos de dispersión
1. Covarianza
1. Coeficiente de correlación (pearson)
1. Coeficiente de determinación $r^2$

```python
plt.hist(df.total_income)
plt.show()
```

```python
 # Correlación entre ingresos y monto del prestamo

# 1. Gráfico de dispersión

sns.scatterplot(data=df,x='total_income',y='loan_amount')
plt.xlim(0, 3e06)
plt.show()
```

```python
df[['total_income', 'loan_amount']].cov()
```

```python
np.var(df.total_income)
```

```python
df[['total_income', 'loan_amount']].corr()
```

```python
# Solo para tarjetas de crédito?
df_tc = df.query("contract_type=='credit card'")
df_tc[['total_income', 'loan_amount']].corr()
```

```python
correlaciones = df[['total_income', 'loan_amount', 'external_score_1', 'external_score_2', 'age','age_mobilephone_days']].corr()
round(correlaciones,2)
```

```python
plt.figure(figsize= (5,3) ) # Ancho de 5 pulgadas y alto de 3 pulgadas
sns.heatmap(correlaciones,annot=True)
plt.show()
```
