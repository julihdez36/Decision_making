### Explorando el performance de los empleados

```python
# Lo siguiente es para evitar alertas molestas al ejecutar el código
import warnings
warnings.filterwarnings(action='ignore')
```

```python
cred_por_empleado = pd.pivot_table(df, index=['Office','Employee'], columns=['year', 'month'], values='loan_amount', aggfunc='count', fill_value=0) # Tabla pivote con la información oficina y empleado a nivel de filas y de año y mes a nivel de columnas
cred_por_empleado = cred_por_empleado.reset_index() # Reset del index para convertirlo en una columna

# Loop para hacer un gráfico por cada oficina
for office in cred_por_empleado['Office', ''].unique():
    df_filtrado = cred_por_empleado[cred_por_empleado['Office', ''] == office].drop('Office', axis=1).set_index('Employee').T # Data frame filtrado, sin la columna oficina y transpuesto
    df_filtrado.plot(figsize=(12,5), title=office) # Gráfico de lineas
```

Cada oficina tiene 3 empleados (excepto 1). Sin embargo se observan brechas enormes en cuanto a la cantidad de créditos gestionados por cada uno de los empleados. Esto muestra enormes diferencias de rendimiento en gestores de crédito que puede ser un problema pues se esperaría que todos los gestores sean buenos y tenga un rendimiento similar.

Por lo tanto en la selección y permanencia de los vendedores pueden estar existiendo problemas que pueden explicar los inconvenientes financieros en los que se encuentra TuCreditoCo.

### Explorando las tasas de impago por oficinas

```python
# Convertir el código de oficina de numérico a string
df['Office'] = df['Office'].astype(str)
```

```python
# Presencia de no pagadores
porcen_nopago = pd.pivot_table(df, index='Office', columns='year', values='default', aggfunc='mean') # Tasa de inpago por oficina y año
porcen_nopago = round(porcen_nopago * 100, 2) # Valores en porcentaje y a dos decimales
porcen_nopago = porcen_nopago.sort_values(by=2019) # Ordenar valores de acuerdo a la tasa de inpago en 2019
porcen_nopago['promedio'] = round(porcen_nopago[2018].mean(),2) # Tasa promedio de inpago en 2018
porcen_nopago # Mostrar la tabla
```

```python
ax = porcen_nopago['promedio'].plot(kind='line', linestyle='dotted', figsize=(12,7), title='Porcentaje de impago') # grafica de linea
porcen_nopago.drop('promedio', axis=1).plot(kind='bar', ax=ax) # Gráfica de barras
```
