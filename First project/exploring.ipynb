```python
# Análisis univariado 

# Variables cualitativas

df.default.value_counts() # frecuencia absoluta
df.default.value_counts(normalize= True) * 100 # frecuencia relativa
```

```python
df.gender.value_counts(normalize=True)
```

```python
# Tabla de contingencia

pd.crosstab(df.gender,df.default, normalize= True, margins= True) * 100
```

```python
sns.countplot(data=df, x= 'gender',color = 'purple')
plt.title('Género de clientes solicitantes de crédito',loc='left')
plt.xlabel = 'Género'
plt.ylabel = 'Frecuencia absoluta (personas)'
plt.grid(color = 'white')
plt.show()
```
